3.

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.title('background_rgb')
plt.show()
```

4.

```python
resize = cv2.resize(rgb, (570, 290))
plt.imshow(resize)
plt.title('background_rgb')
plt.show()
```

5.

```python
rectangle = cv2.line(resize, (4, 4), (4,285), (255, 0, 0), 1)
rectangle = cv2.line(rectangle, (4, 4), (565,4), (255, 0, 0), 1)
rectangle = cv2.line(rectangle, (565, 4), (565,285), (255, 0, 0), 1)
rectangle = cv2.line(rectangle, (4, 285), (565,285), (255, 0, 0), 1)
plt.imshow(rectangle)
plt.title('rectangle')
plt.show()
```

6.

```python
corners = cv2.circle(rectangle, (4, 4), 10, (0, 0, 255), 5)
corners = cv2.circle(corners, (565,4), 10, (0, 0, 255), 5)
corners = cv2.circle(corners, (565,285), 10, (0, 0, 255), 5)
corners = cv2.circle(corners, (4, 285), 10, (0, 0, 255), 5)
plt.imshow(corners)
plt.title('corners')
plt.show()
```
